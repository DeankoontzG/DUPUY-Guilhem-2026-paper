```python
import networkx as nx
import numpy as np
from decorrnet import infer_spatial_decorrelated_louvain_communities, infer_spatial_decorrelated_SiNE_embedding

# 1. Load a NetworkX Graph with a spatial position attribute
G = nx.erdos_renyi_graph(n=50, p=0.1)

# Example : a random 2D position in [0;1] is assigned to every node
for node in G.nodes:
    G.nodes[node]['GT_pos'] = np.random.rand(2)

# 2. Extract spatially decorrelated communities (modifies G in place)
G = infer_spatial_decorrelated_louvain_communities(
    G=G, 
    pos_attr="GT_pos", 
    attr_name="spatially_decorrelated_louvain_id"
)

# 3. Generate spatially decorrelated continuous embeddings (modifies G in place)
G = infer_spatial_decorrelated_SiNE_embedding(
    G=G, 
    pos_attr="GT_pos", 
    attr_name= "spatially_decorrelated_embedding",
    embedding_dim=64
)

# Nodes now contain both decorrelated attributes
print(G.nodes[0])
```
